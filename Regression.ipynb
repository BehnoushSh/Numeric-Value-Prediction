```python
# Regression
import numpy as np
import pandas as pd
from pandas import DataFrame

data=pd.read_csv("/Users/Behnoush/Desktop/Python Projects/Machine learning/Data/train.csv",index_col=0)
data.head(10)

data=DataFrame(data, columns=['MSSubclass','LotFrontage','LotArea','OverallQual','OverallCond','YearBuilt','YearRemodAdd','MasVnrArea','BsmtFinSF1','TotalBsmtSF','BedroomAbvGr','KitchenAbvGr','GarageCars','GarageArea','WoodDeckSF','OpenPorchSF','EnclosedPorch','3SsnPorch','ScreenPorch','PoolArea','MiscVal','MoSold','YrSold','SaleType','SaleCondition','SalePrice'])   

data['SaleCondition']=data['SaleCondition'].replace("Normal",1)
data['SaleCondition']=data['SaleCondition'].replace("Abnorml",0)

data['age']=2019-data.YearBuilt    #Calculate the age of building

data['Renovation']=0               #Calculate a new variable and drop these two columns 
data['Renovation'] = np.where(data['YearBuilt'] == data['YearRemodAdd'],0, 1)

data=data.drop(['YearBuilt','YearRemodAdd','SaleType','MSSubclass'],axis=1)  #'SaleType' can be dropped, no explaination available! 
                                                                             #'MSSubclass' has no value! 
# data['SaleType'].value_counts() 
data.tail(10)
data.describe()
data.dtypes

#Calculate the correlation between the variables(regressors) and dependent variable
data.corr(method='pearson', min_periods=1)    # Here I have the sign of correlations
Cor_value=data.corr(method='pearson', min_periods=1).abs()    # Here I have the absolute values

s = Cor_value.unstack()
Sort_Value = s.sort_values(kind="quicksort")

#Sort_Value.where(Sort_Value>=0.5)  
Sort_Value[Sort_Value>=0.5]    # Could not find how it will be if I had two conditions; e.g. Sort_Value<1 
# These variables have correltion coef above 0.5 with SalePrice:
# Age,TotalBsmtSF,GarageArea,GarageCars,OverallQual

Sort_Value[Sort_Value>=0.4]    # Could not find how it will be if I had two conditions; e.g. Sort_Value<1 
# These variables have correltion coef above 0.4 with SalePrice:
# Age,TotalBsmtSF,GarageArea,GarageCars,OverallQual,MasVnrArea

# I will go with the treshold of 0.5 
# I have 5 predictors and sale price as the dependent variable

data.head()

feature=DataFrame(data,columns=['age','TotalBsmtSF','GarageArea','GarageCars','OverallQual','YrSold'])
X=feature
Y=DataFrame(data,columns=['SalePrice'])    #Y=data.SalePrice
X.head()
Y.head()

from sklearn import linear_model

regr=linear_model.LinearRegression() 
 
# Use the fit command to estimate coefficients
regr.fit(X,Y)

print('Coefficients:', regr.coef_)
print('Intercept:', regr.intercept_)

####################################################################
####################################################################
data_2=pd.read_csv("/Users/Behnoush/Desktop/Python Projects/Machine learning/Data/test.csv",index_col=0)
data_2.head(10)

data_test=DataFrame(data_2, columns=['YearBuilt','YearRemodAdd','TotalBsmtSF','GarageCars','GarageArea','OverallQual','YrSold'])   

data_test['age']=2019-data_test.YearBuilt    #Calculate the age of building
data_test=data_test.drop(['YearBuilt','YearRemodAdd'],axis=1) 

data_test.head(10)
data_test.size

# If you want to drop "Missing values"
X_test = data_test.dropna(axis=0)
X_test.size

# Replace missing values with the mean of variables
#from sklearn.impute import SimpleImputer
#my_imputer = SimpleImputer()
#X_test = my_imputer.fit_transform(data_test)

Est_Y=regr.predict(X_test)
Est_Y.size
df=DataFrame(Est_Y)
df.head(10)

#writer = pd.ExcelWriter('../input/Result.xlsx')
#df.to_excel(writer,'Sheet1')
#writer.save()
```
